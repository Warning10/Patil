```python
import pandas as pd

# Load the dataset into a pandas DataFrame
df = pd.read_csv('iris.csv')
```

```python
column=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
# Calculate the statistics
mean = df[column].mean()
median = df[column].median()
minimum = df[column].min()
maximum = df[column].max()
std_dev = df[column].std()

# Display the statistics
print("Mean:")
print(mean)
print("\nMedian:")
print(median)
print("\nMinimum:")
print(minimum)
print("\nMaximum:")
print(maximum)
print("\nStandard Deviation:")
print(std_dev)
```

```python
# Calculate the statistics grouped by the categorical variable
grouped_stats = df.groupby('Species')['Id'].agg(['mean', 'median', 'min', 'max', 'std', 'count'])

# Display the grouped statistics
print(grouped_stats)
```

```python
df1=df.copy()
df1=df1.drop(['Id'],axis=1)
# Calculate the statistics grouped by the categorical variable
grouped_stats = df1.groupby('Species').describe()

# Display the grouped statistics
print(grouped_stats)
```
