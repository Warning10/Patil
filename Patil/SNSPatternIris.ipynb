```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("iris.csv")
```

```python
df.isnull().any()
```

```python
var = ["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]
 
fig, ax = plt.subplots(2,2, figsize=(9,9))
for var, subplot in zip(var, ax.flatten()):
    sns.boxplot(x='Species', y=var, data=df, ax=subplot)
fig.tight_layout()
plt.show()
```

```python
sns.boxplot(data=df, x="Species", y="SepalLengthCm")
```

```python
# Create a histogram for each feature
plt.figure(figsize=(10, 6))  # Set the figure size

# Iterate over each feature and create a histogram
for feature in df.columns[1:-1]:  # Exclude the 'species' column
    plt.hist(df[feature],rwidth=0.5)
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()
```

```python
df.hist(column='SepalLengthCm',grid=False,figsize=(5,3),rwidth=0.5)
df.hist(column='SepalWidthCm',grid=False,figsize=(5,3),rwidth=0.5)
df.hist(column='PetalLengthCm',grid=False,figsize=(5,3),rwidth=0.5)
df.hist(column='PetalWidthCm',grid=False,figsize=(5,3),rwidth=0.5)
```

```python
sns.histplot(df, x="SepalLengthCm", hue="Species")
```

```python
# 4. Compare distributions and identify outliers
plt.figure(figsize=(10, 6))  # Set the figure size
sns.violinplot(data=df)
plt.title('Violin Plot of Iris Dataset Features')
plt.show()
```
